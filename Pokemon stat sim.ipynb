```python
import math
import random
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
num_set = [str(x) for x in range(1,221)]
rare = [str(x) for x in range(1,71)]
uncommon = [str(x) for x in range(71,141)]
common = [str(x) for x in range(141,221)]
```

```python
num_pack = 10
```

```python
def gotta_catch_em_all():
    set_list = []
    counter = 0
    while len(set_list) < len(num_set):
        working = []
        working = working + random.sample(rare, 1)
        working = working + random.sample(uncommon, 3)
        working = working + random.sample(common, 6)
        for item in working:
            if item not in set_list:
                set_list.append(item)
        counter +=1
    return counter
            
```

```python
num_sims = 1000
plotting_y = []
plotting_x = [x for x in range(1,num_sims+1)]
```

```python
i = 0
while i < num_sims:
    a = gotta_catch_em_all()
    plotting_y.append(a)
    i += 1
    
```

```python
plt.plot(plotting_x,plotting_y)
plt.axhline(y=sum(plotting_y)/len(plotting_y), color='r', linestyle='-')
```

```python
plt.plot(plotting_x,sorted(plotting_y))
plt.axhline(y=sum(plotting_y)/len(plotting_y), color='r', linestyle='-')
```
